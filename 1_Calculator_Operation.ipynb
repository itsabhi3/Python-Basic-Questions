# Calculator


```python
def addition(a,b):
    return a+b

def substraction(a,b):
    return a-b

def multiply(a,b):
    return a*b

def division(a,b):
    return a/b

def floorDivision(a,b):
    return a//b

def power(a,b):
    return a**b

while(1):

    print("Enter Your Choice")
    choice=int(input("1: ADD, 2: Sub, 3: Mul, 4: Div, 5: fDiv, 6: pow, 7:any other key"))

    if choice > 6:
        print("Please choose the right option")
        break
    
    a=int(input("Enter first number"))
    b=int(input("Enter Second number"))
    
    if choice == 1:
        print(addition(a,b))
    elif choice == 2:
        print(substraction(a,b)) 
    elif choice == 3:
        print(multiply(a,b))
    elif choice == 4:
        print(division(a,b))
    elif choice == 5:
        print(floorDivision(a,b))
    elif choice == 6:
        print(power(a,b))
    
    
```
