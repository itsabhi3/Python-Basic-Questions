# Disarium Number
#### 89 and 135 etc, is called Disarium number

```python
def countDigit(n):
    count = 0
    while(n!=0):
        n=n//10
        count+=1
    return count

def disariumNumber(n,c):
    sum =0
    copy_n=n
    while(n!=0):
        sum = sum + (n%10)**c
        c= c-1
        n= n//10
    if (sum == copy_n):
        return True
    return False

n= int(input("Enter any number"))
c= countDigit(n)
print(disariumNumber(n,c))
```
