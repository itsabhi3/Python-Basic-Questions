# Strong Number
#### 1,2,145,40584 etc, is called strong number

```python
def factorial(n):
    if  n==0:
        return 1
    return n*factorial(n-1)

def strongNumber(n):
    copy_n =n
    sum=0
    while(n!=0):
        sum= sum+factorial(n%10)
        n=n//10
    if(sum == copy_n):
        return True
    return False

n=int(input("Enter any number"))
print(strongNumber(n))
    
```
