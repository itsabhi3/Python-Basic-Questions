# Spy Number
#### 22, 123,132,213 etc, is call spy number

```python
def sumOfDigit(n):
    sum=0
    while(n!=0):
        sum =sum+(n%10)
        n=n//10
    return sum

def productOfDigit(n):
    prod=1
    while(n!=0):
        prod =prod*(n%10)
        n=n//10
    return prod

def spyNumber(n):
    if sumOfDigit(n) == productOfDigit(n):
        return True
    return False

n= int(input("enter any number"))
print(spyNumber(n))
```
