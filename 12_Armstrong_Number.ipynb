# Armstrong Number
#### Check this number 153,370,371,407 etc. is called armstrong number

```python
def countDigit(n):
    count =0
    while(n!=0):
        n=n//10
        count+=1
    return count

def armstrong(n,c):
    copy_n =n
    sum=0
    while(n!=0):
        sum = sum + (n%10)**c
        n=n//10
    if sum == copy_n:
        return True
    return False
n= int(input("enter any number"))
c= countDigit(n)
print(armstrong(n,c))
    
```
