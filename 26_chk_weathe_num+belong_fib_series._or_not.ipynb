# Check weather a number belong to fibonacci series or not

```python
def perfectSquare(n):
    sq=n**0.5
    if sq == int(sq):
        return True
    return False

def checkFibonacci(n):
    if perfectSquare(5*(n**2)+4) or perfectSquare(5*(n**2)-4):
        return True
    return False

n =int(input("Enter any number"))
print(checkFibonacci(n))
```
